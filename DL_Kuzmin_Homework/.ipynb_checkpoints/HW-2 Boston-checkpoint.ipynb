```python
import torch
from torch import autograd
from torch.utils.data import TensorDataset, DataLoader
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
filename = ("./housing.csv")
names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
df = pd.read_csv(filename, delim_whitespace=True, names=names)
```

```python
df.head()
```

```python
df.dtypes
```

```python
labels = df.MEDV.values
features = df.drop(['MEDV'], axis=1).values
```

```python
validation_size = 0.20
seed = 7
X_train, X_validation, Y_train, Y_validation = train_test_split(features, labels, test_size=validation_size, random_state=seed)
```

```python
labels = torch.Tensor(Y_train)
features = torch.Tensor(X_train)
```

```python
batch_sz = 10
```

```python
dataset = TensorDataset(features, labels)
data_iter = DataLoader(dataset, batch_sz, shuffle=True)
```

```python
labels_valid = torch.Tensor(Y_validation)
features_valid = torch.Tensor(X_validation)
```

```python
#inferense
dataset_valid = TensorDataset(features_valid , labels_valid)
data_iter_valid = DataLoader(dataset_valid, batch_sz, shuffle=False)
```

```python
for X, y in data_iter:
    print(X.shape, y.shape)
    break
```

```python
model = torch.nn.Sequential(
        torch.nn.BatchNorm1d(13),
        torch.nn.Linear(13, 1)
        )
```

```python
model
```

```python
loss = torch.nn.MSELoss(reduction='mean')
```

```python
optimizer = torch.optim.Adam(model.parameters(), lr=0.0005)
```

```python
num_epohs = 500
for epoch in range(num_epohs):
    for X, y in data_iter:
        optimizer.zero_grad()
        y_hat = model.forward(X).reshape(-1) 
        l = loss(y_hat, y)
        l.backward()
        optimizer.step()
    l = loss(model.forward(features).reshape(-1),labels)
    print('epoch %d, loss_train: %f' % (epoch, l.item()))
    for X, y in data_iter_valid:
        y_hat = model.forward(X).reshape(-1)
        l = loss(y_hat, y)
    l = loss(model.forward(features_valid).reshape(-1),labels_valid)
    print('epoch %d, loss_valid: %f' % (epoch, l.item()))
```
