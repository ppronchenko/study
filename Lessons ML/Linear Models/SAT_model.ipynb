# SAT model

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import  linear_model
```

```python
data = pd.read_csv('sat.txt', sep = ' ')
data.head()
```

### Строим зависимость univ_GPA от high_GPA

В итоге необходимо получить что-то вроде такого:

```python
from sklearn.linear_model import LinearRegression
Y = data.loc[ :, ['univ_GPA']]
X = data.loc[ :, ['high_GPA', 'math_SAT', 'verb_SAT', 'comp_GPA']]
model = LinearRegression()
model.fit(X, Y)
y_predict = model.predict(X)
plt.plot(Y, y_predict, 'o')
#plt.plot(Y, y_predict, 'k-')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
Y = data.loc[ :, ['univ_GPA']]
X = data.loc[ :, ['math_SAT']]
model = LinearRegression()
model.fit(X, Y)
y_predict = model.predict(X)
plt.plot(X, Y, 'o')
plt.plot(X, y_predict, 'k-')
plt.show()
```
