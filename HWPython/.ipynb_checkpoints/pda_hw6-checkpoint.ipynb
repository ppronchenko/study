```python
import pandas as pd
```

```python
df = pd.read_csv('/home/pavel/HomeWork/study/LessonsPython/keywords.csv')
```

```python
geo_data = {
 
    'Центр': ['москва', 'тула', 'ярославль'],
 
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
 
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
 
}
```

```python
geo_data['Центр']
```

```python
def geo_classified (row):
    for key in geo_data.keys():
        for i in geo_data[key]:
            if i in row['keyword'].lower():
                return key
            else:
                return 'undefined'
```

```python
df['region'] = df.apply(geo_classified, axis=1)
```

```python
df['region'] .value_counts()
```

```python
ratings = pd.read_csv('/home/pavel/HomeWork/study/LessonsPython/ml-latest-small/ratings.csv')
```

```python
def ratings_classified(row):
    if row['rating'] <=2:
        return 'низкий рейтинг'
    elif row['rating'] > 2 and row['rating'] <= 4:
        return 'средний рейтинг'
    elif row['rating'] >= 4.5 and row['rating'] <= 5:
        return 'высокий рейтинг'
    else:
        return 'рейтинг не определен'    
```

```python
ratings['class'] = ratings.apply(ratings_classified, axis=1)
```

```python
ratings['class'].value_counts()
```

```python
ratings_count = ratings.groupby('userId').count().reset_index()[['userId', 'movieId']]
```

```python
film_fans_ratings_count = ratings_count[ ratings_count['movieId'] >= 100 ]
```

```python
film_fans_user_ids = film_fans_ratings_count['userId'].tolist()
```

```python
fans_data = ratings[ ratings['userId'].isin(film_fans_user_ids) ]
```

```python
fans_data_lifetime = fans_data.groupby('userId').agg({'timestamp': ['min', 'max']}).head()
```

```python
fans_data_lifetime['lifetime']= fans_data_lifetime['timestamp']['max'] - fans_data_lifetime['timestamp']['min']
```

```python
fans_data_lifetime['lifetime'] = fans_data_lifetime['lifetime']/3600/24
```

```python
years = list(range(1950, 2011))
```

```python
def production_year(row):
    for i in years:
        if str(i) in row['title']:
            return str(i)
    return '1900'
```

```python
movies = pd.read_csv('/home/pavel/HomeWork/study/LessonsPython/ml-latest-small/movies.csv')
```

```python
movies['year_of_production'] = movies.apply(production_year, axis = 1)
```

```python
joined = ratings.merge(movies, on='movieId', how='left')
joined.head()
```

```python
avg_rating = joined.groupby('year_of_production').agg({'rating': 'mean'})
```

```python
avg_rating.sort_values('rating', ascending=False)
```
