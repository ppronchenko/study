```python
gasoline_engine = auto[auto['fuel-type']=='gas']
diesel_engine = auto[auto['fuel-type']=='diesel']
```

```python
stats.ttest_ind(gasoline_engine['wheel-base'], diesel_engine['wheel-base']) 
```

```python
stats.ttest_ind(gasoline_engine['price'], diesel_engine['price']) 
```

```python
audi = auto[auto['make']=='audi']
bmw  = auto[auto['make']=='bmw']
```

```python
stats.ttest_ind(audi['price'], bmw['price']) 
```

```python
stats.ttest_ind(audi['wheel-base'], bmw['wheel-base']) 
```
